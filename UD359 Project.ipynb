# Answering the question

## Having a look at the data

Now we are going to import the data and have a look at it:

```python
import pandas as pd
print 'Pandas version ' + pd.__version__
subway = pd.read_csv('https://raw.githubusercontent.com/phoenixza1/ud359/master/data/turnstile_data_master_with_weather.csv')
```

```python
print list(subway.columns)
print str(len(list(subway.columns))) + ' columns in total'
```

As we can see, we have 14 columns dedicated to weather, 2 of those to do specifically with rain. In addition 2 columns describe subway usage, 3 time or date and the last one, 'UNIT' is the particular station (we can also refer to the <a href='http://web.mta.info/developers/resources/nyct/turnstile/ts_Field%20Description.txt'>MTA reference</a> for what some of these columns mean).


Let's begin deciding which of these columns we are going to use and which we can get rid of. Let's start with 'ENTRIESn_hourly' and 'EXITSn_hourly', otherwise known as the cumulative number of people entering or exiting a station up to that point in time (then resetting to 0 for the next reading). All we need to do is track how many people go through a station and it doesn't matter whether they are coming or going. So to help us decide which one to use, we are going to see what they look like compared to one another by graphing all Entries and Exits for the entire month together against the Date:

```python
subway['DATEn'] = pd.to_datetime(subway['DATEn'])
subway_plot = subway.groupby('DATEn').sum().reset_index()

from ggplot import *
%matplotlib inline

print ggplot(subway_plot, aes(x = 'DATEn')) +\
               geom_line(aes(y = 'ENTRIESn_hourly'), color = 'blue') +\
               geom_line(aes(y = 'EXITSn_hourly'), color = 'red') +\
               ggtitle('Number of total entries (blue) and exist (red) as a function of date') +\
               xlab('Date') + ylab('Entries or Exists') + theme_bw()
                
print 'Total difference between Entries and Exits: '+ str(round(subway['ENTRIESn_hourly'].sum() - subway['EXITSn_hourly'].sum(),-6))
```

Either about 28 million people became trapped in the subway system during the month of May, or the turnstile exists data is incomplete... (it's the latter).

We can see that the exits curve is virtually the exact same shape, so it could be just as usfeul as the entries data even though it has a large systematic error (assuming its not entries with the problem), which could be the result of a consistent percentage of riders avoiding exit turnstiles, or certain stations not having any. We will use the Entries columns simply due to it's larger numbers which could give more pronounced results from analyses.

Now let's have a look at some of the weather data for a particular station on a particular day (R001 and 2011-05-02): 

```python
subway_usage = subway[['UNIT','DATEn','TIMEn','maxpressurei', 'maxdewpti', 'mindewpti', 'meandewpti', 'fog', 'rain', 'precipi']]
subway_usage = subway_usage[subway_usage['UNIT'] == 'R001']
subway_usage = subway_usage[subway_usage['DATEn'] == '2011-05-02']
subway_usage
```

All these values look suspiciously identical to each other...
Lets see how many unique values there are for some weather columns for all the stations <b>combined</b> during an entire (and different - 2011-05-10) day:

```python
date = '2011-05-02'
subway_usage = subway[subway['DATEn'] == date]
print 'Number of maxpressurei unique values for all UNITs on {}: {}'.format(date,len(subway_usage['maxpressurei'].unique()))
print 'Number of maxdewpti unique values for all UNITs on {}: {}'.format(date,len(subway_usage['maxdewpti'].unique()))
print 'Number of precipi unique values for all UNITs on {}: {}'.format(date,len(subway_usage['precipi'].unique()))
```

It seems that the weather data is all very coarse, in the sense that there is really only one weather entry shared for every station for an entire day.

What this means is that there is no point taking into account different times of day or even different stations. Instead we are going to drop the unit, hour and time columns (as well as 'Unnamed: 0', 'DESCn', exits and 'thunder' (as there is never any)) and group the rest by date, aggregating the entries column by summing it and the others by just selecting their only unique values. This will produce a dataset of only 30 entries for each of the 15 columns:

```python
import numpy as np
## Columns we'll keep
columns = ['DATEn', 'ENTRIESn_hourly', 'maxpressurei', 'maxdewpti', 'mindewpti', 'minpressurei',\
           'meandewpti', 'meanpressurei', 'fog', 'rain', 'meanwindspdi', 'mintempi', 'meantempi', 'maxtempi', 'precipi']
## Group all by date
subway_grouped = subway[columns].groupby('DATEn')
## Aggregating function that just gives first (only) unique value
func = lambda x: x.unique()[0]
## Make a dictionary that assigns that function to all columns but date and entries
agg_dic = dict([[columns[i], func] for i in range(2, len(columns))])
## Add entries to that with a sum function instead
agg_dic['ENTRIESn_hourly'] = np.sum
## Now aggregate the grouped DF as the dictionary specifies
subway_grouped = subway_grouped.agg(agg_dic).reset_index()
print 'Data-Frame shape: ' + str(subway_grouped.shape)
```

## Finding important features for our model

What we need to do now is find the columns (or what we will now call features for the purpose of modelling) that have the greatest effect on ridership. We will assume that the values of all these features are independent of what the entries are.

Before we check out the weather features, let's first investigate something else about the ridership we discovered in the first graph, that is that the ridership reduces significantly during the weekend and on the 30th (Memorial Day). This would therefore be an extremely important feature in our model (it's not the weather yet I know) and we must take it into account when investigating the weather features, because if we don't we could for example, overestimate the effect a weather feature has on ridership simply because it occurs on a holiday. 

So what we will do now is create a new feature called isworkday (we will use the same function on the original data too for the model), where 1 is a workday and 0 is not and display statistics proving the differences:

```python
## make function that returns a series where in 1 is a workday, else 0
def isworkday(dates):
    isworkday_lis = list()
    for date in dates:
        ## Day 30th is memorial day and weekday 6 and 7 is Saturday and Sunday respectfully.
        if date.day != 30 and date.isoweekday() not in [6,7]:
            isworkday_lis.append(1)
        else:
            isworkday_lis.append(0)
    return pd.Series(isworkday_lis)

## Create new column using function
subway_grouped['isworkday'] = isworkday(subway_grouped['DATEn'])

## Create a filtered dataframe for later use
subway_grouped_workdays = subway_grouped[subway_grouped['isworkday'] == 1]
subway_grouped_offdays = subway_grouped[subway_grouped['isworkday'] == 0]

## Show statistics
workdays_test = pd.DataFrame()
workdays_test['Ridership on Workdays'] = subway_grouped_workdays['ENTRIESn_hourly'].describe()
workdays_test['Ridership on Offdays'] = subway_grouped_offdays['ENTRIESn_hourly'].describe()
workdays_test
```

We can clearly see the differences in ridership caused by off days, particularly the mean, min and max.

Now lets plot the 4 different types of numeric weather features (Pressure, Dew point, Temperature and Precipitation with blue dots) against ridership for all work days and a statistically smoothed line (black) in order to determine if there is a correlation between them and ridership:

```python
plot1 = ggplot(subway_grouped_workdays, aes(y = 'ENTRIESn_hourly', x = 'meanpressurei')) +\
                geom_point(aes(x = 'meanpressurei'), color = 'blue') +\
                stat_smooth(span=1) + theme_bw() + xlab('Mean Pressure') + ylab('Ridership') +\
                ggtitle('Ridership v Mean Pressure')  
        
plot2 = ggplot(subway_grouped_workdays, aes(y = 'ENTRIESn_hourly', x = 'meandewpti')) +\
                geom_point(aes(x = 'meandewpti'), color = 'blue') +\
                stat_smooth(span=1) + theme_bw() + xlab('Mean Dew Point') + ylab('Ridership') +\
                ggtitle('Ridership v Mean Dew Point')  
        
plot3 = ggplot(subway_grouped_workdays, aes(y = 'ENTRIESn_hourly', x = 'meantempi')) +\
                geom_point(aes(x = 'meantempi'), color = 'blue') +\
                stat_smooth(span=1) + theme_bw() + xlab('Mean Temperature') + ylab('Ridership') +\
                ggtitle('Ridership v Mean Temperature')  
        
plot4 = ggplot(subway_grouped_workdays, aes(y = 'ENTRIESn_hourly', x = 'precipi')) +\
                geom_point(aes(x = 'precipi'), color = 'blue') +\
                stat_smooth(span=1) + theme_bw() + xlab('Percipitation') + ylab('Ridership') +\
                ggtitle('Ridership v Precipitation')

plot1, plot2, plot3, plot4
```

As we can see, there is no clear relationship (the points are either all over the place or there are too few points outside a central cluster to establish a pattern) between any of these features and ridership.

Now lets have a look at rainy work days versus non rainy work days, which should have the largest effect and what we are most interested in:

```python
subway_plot = subway_grouped_workdays.copy()
subway_plot['Ridership Mean'] = subway_plot['ENTRIESn_hourly']
subway_plot['Ridership SD'] = subway_plot['ENTRIESn_hourly']
subway_plot = subway_plot.groupby('rain')
subway_plot = subway_plot.agg({'Ridership Mean': np.mean,'Ridership SD': np.std})
subway_plot = subway_plot.reset_index()
subway_plot[['rain','Ridership Mean','Ridership SD']]  
```

The two samples seem very similar to each other based on their means and standard deviations. To be sure as to whether or not there is a statistically significant difference between them, we are going to need to use either Welsh's t-test or Mann Whitney's U-test. 

Which one we use depends upon whether or not the data is normally distributed, so lets determine that now by graphing it and using the Shapiro-Wilk test. Remember, the following is ridership data for work days regardless of rain:

```python
## Graph
print ggplot(subway_grouped_workdays, aes(x = 'DATEn', y = 'ENTRIESn_hourly')) +\
               geom_point(color = 'blue') + stat_smooth(span=1) +\
               ggtitle('Number of total entries as a function of date') +\
               xlab('Date') + ylab('Entries') + theme_bw()
            
## S-W statistic
import scipy
import scipy.stats
p = scipy.stats.shapiro(subway_grouped_workdays['ENTRIESn_hourly'])[1]
print 'Shapiro-Wilk P-value: {}'.format(p)
```

Judging by the graph, it's not normally distributed and the Shapiro-Wilk test confirms this. With the null hypothesis for the test being that the data is normally distributed, we can clearly see that with a P-value of 0.0017 (and a chosen alpha of 0.05), we can definitely reject that null and say that the data is <b>not</b> normally distributed.

Therefore, we will use the Mann Whitney U-test:

```python
with_rain = subway_grouped_workdays[subway_grouped_workdays["rain"] == 1]['ENTRIESn_hourly']
without_rain = subway_grouped_workdays[subway_grouped_workdays["rain"] == 0]['ENTRIESn_hourly']
U, P = scipy.stats.mannwhitneyu(with_rain, without_rain)
n_rain = len(with_rain)
n_norain = len(without_rain)
med_rain = with_rain.median()
med_norain = without_rain.median()

print "Median values for rain and no rain is {} and {} respectivel\
y".format(med_rain,med_norain)
print "There was no significant difference between the samples,"
print "(Mann–Whitney U = {}, n_rain = {}, n_norain = {}, P = {}, P \
> 0.05 two-tailed)".format(U,n_rain,n_norain,P)
```

Based on those results, we cannot reject the Null hypothesis. However, we can see that the sample sizes where 7 and 13 which raises several problems, namely that the two samples sizes are significantly different from each other and should ideally be identical to each other. Also, the test compares the medians of the samples to each other and so we cannot definitively say that the samples are the same to each other. Finally the sample sizes are quite small, where sizes of both 20 or more would yield more useful results.

Another limitation in general with the rain value, is that it's essentially a boolean value and does not take into account how much rain fell, however the Ridership v Precipitation graph from earlier didn't show a correlation either.

## Machine Learning

We are going to use a supervised machine learning algorithm in order to quantitatively show a correlation (or lack thereof) between weather features and ridership through the power of prediction. We are using supervised learning because we want to keep the data-points labelled and thus in context. We are going to use linear regression because we are looking for a linear relationship between features and ridership, the features are independent of one another, we know their values precisely and ridership is not derived from any of the features.

We are going to use Ordinary Least squares, the most straight forward linear regression and not gradient descent, as the data set is not large enough to warrant it. We are going to do multiple regressions and show their corresponding R^2 values. All the regressions will use isworkday, Hour and the UNIT as features. It is impossible to ignore the latter feature and obtain an acceptable R^2 regardless of which other features are included. This first regression will have all the weather features except 'rain' because it is dependant on 'precipi' and the latter is more accurate:

```python
## Done to represent units in a way that can be modeled.
unit_horizontal = pd.get_dummies(subway['UNIT'], prefix='unit')

## Features
features_list = ['maxpressurei', 'maxdewpti', 'mindewpti', 'minpressurei', 'meandewpti', 'meanpressurei',\
                 'fog', 'meanwindspdi', 'mintempi', 'meantempi', 'maxtempi', 'precipi', 'Hour']
features = subway[features_list].join(unit_horizontal)
features['isworkday'] = isworkday(subway['DATEn'])
features_array = np.array(features)

## Values
values = subway[['ENTRIESn_hourly']]
values_array = np.array(values).flatten()

## Regression
thetas = np.linalg.lstsq(features_array, values_array)[0]
predictions = np.dot(features_array, thetas)

r2_1 = 1 - np.square(values_array - predictions).sum() / np.square(values_array - np.mean(values_array)).sum()

print 'R^2_1: ' + str(r2_1)
```

Here is the R^2 score for a regression with no weather features used:

```python
## Done to represent units in a way that can be modeled.
unit_horizontal = pd.get_dummies(subway['UNIT'], prefix='unit')

## Features
features_list = ['Hour']
features = subway[features_list].join(unit_horizontal)
features['isworkday'] = isworkday(subway['DATEn'])
features_array = np.array(features)

## Values
values = subway[['ENTRIESn_hourly']]
values_array = np.array(values).flatten()

## Regression
thetas = np.linalg.lstsq(features_array, values_array)[0]
predictions = np.dot(features_array, thetas)

r2_2 = 1 - np.square(values_array - predictions).sum() / np.square(values_array - np.mean(values_array)).sum()

print 'R^2_2: ' + str(r2_2)

print '% Difference between R^2_1 and R^2_2: ' + str(round(100 - r2_1 / r2_2 * 100,1))
```

As we can see, removing the weather features only decreased the R^2 score by about 0.1%.

There are several potential problems with using OLS and linear regression in general: Firstly, many of the features (such as Hour) where not linearly correlated with ridership (under-fitting). Secondly in the first regression there may have been too many features, which would lead the model to try and fit the 'noise' generated by all the features (over-fitting) and another problem caused by too many features is that the R^2 score will always increase with the number of features added, which could lead to the false assumption that the weather features even improved the model at all, when it in fact may have reduced its accuracy.

# Ways to improve the study

Because there is essentially only 30 data-points for each weather feature (one for each day of the month), we would have benefited greatly if either weather measurements for each hour of the day where included and/or more than one month's worth of data was studied. Because we didn’t have either of these, we had too few data-points to explore potential relationships and our statistical tests and models (namely U-test) had too small a sample to be considered reliable.

## Map-Reduce

But we must be careful what we wish for, because as we pointed out in the introduction we live in a world filled with data. If we where to use far more data in this study, we would end up using far more computational power and even exceed the capacity of a single computer (particularly storage space and RAM). Granted, the source data used in this study was only 16MB and that we would need 312500x that in order to fill 5TB worth storage space, we could however quickly approach that when we add much more weather data, more frequent intervals for all data and longer time spans. Plus analyses would take an exceedingly long time with only one machine.

So to solve this potential problem, we could use Map-Reduce to distribute computational requirements across as many systems as we need, splitting up the data, sending the fragments off to different mappers to find out which reducer to send specific parts of the data to and process and aggregate the data there. The data however, would initially be difficult to discover (and figure out what to map and how to reduce), as Map-Reduce is not as easy to work with due to its necessary complexity. Several tools exist that could help with this, namely Giraph for graphing the data, Mahout for using machine learning to explore the data and finally Pig to provide a layer of abstraction to help us with conceptualizing the data and problems. All of these Map-Reduce infrastructures are built on top of Hadoop.
